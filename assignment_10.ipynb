```python
book_df = df.pivot(index='userid',columns='booktitle',values='rating').reset_index(drop=True)
```

```python
book_df
```

```python
book_df.index = df.userid.unique()
```

```python
book_df
```

```python
book_df.fillna(0, inplace=True)
```

```python
book_df
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine
```

```python
user_sim = 1 - pairwise_distances(book_df.values,metric='cosine')
```

```python
user_sim
```

```python
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df.index = df.userid.unique()
user_sim_df.columns = df.userid.unique()
```

```python
user_sim_df.iloc[0:10, 0:10]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:10, 0:10]
```
